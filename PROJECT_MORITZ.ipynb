```python
# permutation test for checking if there is a significant reletionship between time to delivery and  returned orders
complete=df_definitivo[df_definitivo['status'] == 'Complete']['time_to_complete'].dropna().to_list()
returned=df_definitivo[df_definitivo['status'] == 'Returned']['time_to_complete'].dropna().to_list()
```

- $H_0: \mu_{complete} = \mu_{returned}$
- $H_1: \mu_{returned} > \mu_{complete}$

```python
print(len(complete))
print(len(returned))
```

```python
import random
```

```python
complete=random.sample(complete, 18208)
```

```python
mu_complete = np.mean(complete)
mu_returned = np.mean(returned)
mu_diff = mu_complete - mu_returned
print(mu_diff)
```

```python
df_permutation = pd.DataFrame({'group': ['complete' if el <= len(complete) else 'returned' for el in range(1,(len(returned)*2)+1)], 
                   'responses': complete + returned})
df_permutation.sample(5)
```

```python
n = 20000
mu_diff_samples = []
for i in range(n): 
    df_sample1 = df_permutation.sample(frac=0.5, replace=False)
    ix2 = set(df_permutation.index) - set(df_sample1.index)
    df_sample2 = df_permutation.iloc[list(ix2)]
    mu_diff_temp = np.mean(df_sample1['responses']) - np.mean(df_sample2['responses'])
    mu_diff_samples.append(mu_diff_temp)
plt.figure(figsize=(8, 6))
sns.histplot(mu_diff_samples)
plt.axvline(mu_diff, 0, 1, color='r', linestyle='--')
plt.show()
print('mu_diff:', round(mu_diff, 2))
```

```python
print('Results more extreme than observed:' ,sum(mu_diff_samples>mu_diff))
print('Total results:', len(mu_diff_samples))
p_value=sum(mu_diff_samples>mu_diff)/n
print('p-value: ', p_value)

alpha = 0.05
print('alpha:', alpha)
if p_value <= alpha: 
    print('p-value <= alpha | the observed result is statistically significant | H0 rejected')
else: 
    print('p-value > alpha | the observed result is not statistically significant | H0 accepted') 
```

```python
# There is no significant association between delivery time and returned status.
```
